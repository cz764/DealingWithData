#### Using the map command

The map command in Python has the following syntax:

`map(function, [list of values for first argument], [list of values for second argument]...)`

It takes as input a function, which has a set of parameters. Then, it iterates over the lists that follow; the lists contain the arguments that are passed to the function. Map returns a list of values that are the result of applying the function to all the elements of the list(s). 

For example, in the following code, the `add` function is going to be applied to the two lists (`[1, 2, 3, 4]` and `[9, 10, 10, 11]`) that follow. The result of the map will be a list containing the values `[add(1,2), add(2,10), add(3,10), add(4,11)]`




```python
def add(x,y):
    return x+y
    
example = map(add, [1, 2, 3, 4], [9, 10, 10, 11])
example
```

#### Using the map for dataframes

Using the map command, we can:
* Create new columns for the dataframe
* Modify existing columns
* Generate new lists that are the result of operations on the columns of the dataframe

For example, suppose that Pandas were not able to understand that `INSPDATE` is a datetime. We can apply the `to_datetime` function to all the elements of that column, and replace the column with datetime objects:

```python
from pandas.tseries.tools import to_datetime

restaurants['INSPDATE'] = map(to_datetime, restaurants['INSPDATE'])
```

```python
restaurants
```

Or that we really do not like the encoding of boroughs in the "BORO" attribute, so we want to replace them with their real values. We can define a function that takes as input the borough code, and returns the proper name of the borough. Then, we use the map function to apply it to the column "BORO". We can then either replace the "BORO" column with the revised version, or (more safe) create a new column with the more readable version:

```python
def boroToString(borocode):
    if borocode == "1":
        return "Manhattan"
    elif borocode == "2":
        return "The Bronx"
    elif borocode == "3":
        return "Brooklyn"
    elif borocode == "4":
        return "Queens"
    elif borocode == "5":
        return "Staten Island"
    else:
        return "N/A"
    
restaurants['Borough'] = map(boroToString, restaurants['BORO'])
```

```python
restaurants[["DBA", "BORO", "Borough"]]
```

### In class excercise: using Map

```python
def postalAddress(restaurantName, strNo, strName, borough, zipcode):
    address = "%s\n%s %s\n%s, NY %s" % (restaurantName, strNo, strName, borough, zipcode)
    return address 
                                    
restaurants['Address'] = map(postalAddress, restaurants['DBA'], restaurants['BUILDING'], restaurants['STREET'], restaurants['Borough'], restaurants['ZIPCODE'])
```

```python
restaurants[["DBA", "Address"]]
```

### Selecting rows

To select rows, we can use the following approach, where we generate a list of boolean values, one for each row of the dataframe, and then we use the list to select which of the rows of the dataframe we want to keep"

```python
is_08A = (restaurants["VIOLCODE"] == "08A")
inspections08A = restaurants[is_08A]
inspections08A["DBA"].value_counts()[:10]
```

And we can use more complex conditions:

```python
is_08A_manhattan = (restaurants["VIOLCODE"] == "08A") & (restaurants["BORO"] == "1")
inspections08A_in_manhattan = restaurants[is_08A_manhattan]
inspections08A_in_manhattan["DBA"].value_counts()[:10]
```

```python
is_08A = (restaurants["VIOLCODE"] == "08A")
violations08A = restaurants[is_08A]
violations08Acounts = violations08A['BORO'].value_counts()
boro_counts = restaurants['BORO'].value_counts()
```

```python
violations08Acounts / boro_counts
```

```python
(violations08Acounts / boro_counts).plot(kind='bar')
```
